```python
df1 = data1.dropna(axis=1, how='all')
```

```python
df1 = df1.fillna(0)
```

```python
indicator_cols = ['1960', '1961', '1962', '1963', '1964', '1965', '1966', '1967', '1968',
                  '1969', '1970', '1971', '1972', '1973', '1974', '1975', '1976', '1977',
                  '1978', '1979', '1980', '1981', '1982', '1983', '1984', '1985', '1986',
                  '1987', '1988', '1989', '1990', '1991', '1992', '1993', '1994', '1995',
                  '1996', '1997', '1998', '1999', '2000', '2001', '2002', '2003', '2004',
                  '2005', '2006', '2007', '2008', '2009', '2010', '2011', '2012', '2013',
                  '2014', '2015', '2016']

# Determine which rows have at least one non-zero value in the indicator columns
non_zero_rows = (df1[indicator_cols] != 0).any(axis=1)
df_non_zero = df1[non_zero_rows]
```

```python
print("Number of columns before processing: ",len(df_non_zero.columns))
```

```python
df1 = df_non_zero
```

Q2:

```python
# Filter DataFrame to include only the five countries of interest
countries = ['India', 'United States', 'China', 'United Kingdom', 'France']
df_filtered = df1[df1['Country Name'].isin(countries)]

# Select relevant columns
df_co2_2012 = df_filtered.loc[:, ['Country Name', '2012', 'Indicator Code']]

# Calculate the mean value of CO2 emissions in 2012 for the five countries
mean_co2_2012 = df_co2_2012.loc[:, '2012'].mean()

print("Average CO2 emissions in 2012 for India, USA, China, United Kingdom, and France: {:.2f}".format(mean_co2_2012))
```

```python
# Filter DataFrame to only include CO2 emissions data for 2016
df_co2_2016 = df1[(df1['Indicator Code'] == 'EN.ATM.CO2E.KT') & (df1['2016'].notnull())]

# Calculate the total CO2 emissions for 2016
total_co2_2016 = df_co2_2016['2016'].sum()

# Calculate proportional CO2 emissions for each country on 2016
df_co2_2016['weights'] = df_co2_2016['2016'] / total_co2_2016 * 100

# Determine top 5 countries that emit the maximum CO2 on 2016
top5_co2_2016 = df_co2_2016.sort_values(by='2016', ascending=False).head(5)
top5_co2_2016

# Display the results
print("Top 5 countries with the maximum CO2 emissions on 2016:")
print(top5_co2_2016.loc[:, ['Country Name', '2016']])
```

```python
temp = df1.set_index("Country Name")


max_year_sum = -1
max__year = -1


for column in temp.columns[4:]:
    year_sum = sum(temp.loc[["India" , "United States" , "United Kingdom" , "China" , "France"]][column])
    if year_sum > max_year_sum:
        max_year_sum = year_sum
        max__year = column


max__year
```

```python
sum_2016 = sum(df1['2016'])


df1['weights'] = (df1['2016'] * 100) / sum_2016
df1.sort_values(by='weights', ascending=False)[:5][['Country Name' , 'weights']]
```

```python
population = pd.read_csv("Population.csv")
```

```python
temp = df1[:]
pop_temp = population[['Country Name' , '2016']].set_index('Country Name')
weights_per_person = []


for i in range(df1.shape[0]):
    weights_per_person.append(float (df1.iloc[i]['2016'] / pop_temp.loc[df1.iloc[i]['Country Name']] ))
sum_weights_per_person = sum(weights_per_person)


weights_per_person = np.array(weights_per_person)


df1['weights_per_person'] = (weights_per_person *  100) / sum_weights_per_person
df1.sort_values(by='weights_per_person', ascending=False)[:5][['Country Name' , 'weights_per_person']]
```
